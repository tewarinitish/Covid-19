```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# dataframe reads

df_india = pd.read_csv("India_record_update.csv")
df_italy = pd.read_csv("Italy_record_update.csv")
df_china = pd.read_csv("China_record_update.csv")
df_us = pd.read_csv("US_record_update.csv")
```

```python
# taking only significant weeks of the countries
# note: India 25-02 onward, Italy & US 11-02 onward, china all
# also, I initially considered only communicable cases i.e subtracting deaths + recovered from confirmed cases
# since medical reports in china claimed to have brought down the cure rate by a lot,
# it could cause inconsistency in our dataset, so I decided to only include the confirmed cases
# I have trained two different models on just china dataset (active cases only)
# and all three combined with total number of cases

X_india = df_india["Confirmed"] #- df_india["Deaths"] - df_india["Recovered"]
X_india = X_india[5:]

X_italy = df_italy["Confirmed"] #- df_italy["Deaths"] - df_italy["Recovered"]
X_italy = X_italy[3:]

X_china = df_china["Confirmed"] - df_china["Deaths"] - df_china["Recovered"]

X_us = df_us["Confirmed"] #- df_us["Deaths"] - df_us["Recovered"]
X_us = X_us[3:]

Xs = [X_italy, X_china, X_us]
Xs = [X_china]
ys = X_india
```

```python
# preparing training and test sets solely based on number of cases vs day n

X_train = []
y_train = []

for X in Xs:
    day = 0
    for cases in X:
        X_train.append([day])
        y_train.append(cases)
        day += 1
        
X_test = []
y_test = []

day = 0
for cases in ys:
    X_test.append([day])
    y_test.append(cases)
    day += 1
```

```python
# this is a plot of only china's data and it shows how they have flattened the curve after a certain number of cases

plt.scatter(X_train, y_train, color='blue', linewidth=1)
plt.show()
```

```python
# selecting n polynomial features; 4 turned out to be the best in this case

poly = PolynomialFeatures(4)
```

```python
# fitting linear regression model but with polynomial features and predicting

reg = LinearRegression().fit(poly.fit_transform(X_train), y_train)
y_pred = reg.predict(poly.fit_transform(X_test))
```

```python
# india (black dots) has not arrived at china's condition so we do not have enough data on india, yet
# to be able to tell whether the immediate lockdown is going to be effective, we need more features like exact spans of lockdown
# that the other countries implemented

plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```

```python
# this is a scatter plot of all accounted cases of the test set countries

plt.scatter(X_train, y_train, color='blue', linewidth=1)
plt.show()
```

```python
# this shows that india (the black dots)
# is actually doing way better compared to the other three countries (blue line; an average trained model)
# however, this is only the number of cases that have been REPORTED across the states so we still have to play safe
# this also shows that we do not have enough data in this dataset 

plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
