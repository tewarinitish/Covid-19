```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import datetime
```

```python
file = pd.read_csv('China_record_update.csv')
file1 = pd.read_csv('US_record_update.csv')
file2 = pd.read_csv('India_record_update.csv')
```

```python
plt.figure(figsize=(10, 5))
#sns.lineplot(data=acc_data_r_d)
plt.plot(file['Date'],file['Confirmed'])
plt.plot(file['Date'],file['Deaths'])
plt.plot(file['Date'],file['Recovered'])
plt.title("China cases over time", fontsize=20)
plt.xlabel("Date", fontsize=10, size=15)
plt.ylabel("Values", fontsize=10, size=15)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
#plt.legend(loc="upper left",fontsize=20)
plt.legend(['Confirmed', 'Deaths' , 'Recovered'],loc="upper left",fontsize=10)
plt.show()
```
